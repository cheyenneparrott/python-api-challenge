### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Configure Google Maps Key
gmaps.configure(api_key=g_key)

humidity_df = cities_df["Humidity"]

locations_df = cities_df [["Lat" , "Lon"]]
```

```python
#Add Heatmap

heatmap_df = gmaps.figure(center = (20, 0), zoom_level = 2.2)
heatlayer_df = gmaps.heatmap_layer(locations_df, humidity_df, dissipating = False, max_intensity = 100, point_radius = 5)
heatmap_df.add_layer(heatlayer_df)
heatmap_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# hotel_df = criteria_df
# hotel_df.head()
s
criteria_df["Hotel Name"] = ""
hotel_df = criteria_df[["City", "Country", "Hotel Name", "Cloudiness", "Lat", "Lon", "Max Temp", "Wind Speed"]]
```

```python
# API Query
# Hotel_List = []
g_key = "AIzaSyAZAFWveZlatKaTyHMgeDCakGukecdCag0"
# Create Parameters dictionary
params_df = {"radius": 5000, "types": "hotel", "key": g_key}

for index, row in hotel_df.iterrows():
    # Get Lat and Long from hotel_df
    lat_value = row["Lat"]
    long_value = row["Lon"]
    params_df["location"] = f"{lat_value},{long_value}"

    final_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # Get weather data
    weather_response = requests.get(final_url, params = params_df)
    weather_json = weather_response.json()
    
    try:
        hotel_df.loc[index, "Hotel Name"] = weather_json['results'][0]["name"]
                      
        print(f" The name of the hotel is {weather_json['results'][0]['name']}")
    except(KeyError, IndexError):        
        print(f" Hotel name not found.")

```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lon"]]
```

```python
# Add marker layer ontop of heat map
fig = gmaps.figure()
marker_layer_df = gmaps.marker_layer(locations, info_box_content = hotel_info)
fig.add_layer(marker_layer_df)

# Display figure
fig
```
